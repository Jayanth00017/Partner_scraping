```python
target_stud=[]
for i in urls:
    try:
        driver.get(i)
        pr=(driver.find_element_by_class_name("egCard__item"))
        target_stud.append(pr.find_element_by_tag_name("ul").text)
        
        print("\n\n",i,"\n")
    except:
        target_stud.append(" ")

target_students=[]
for i in target_stud:
    try:
        target_students.append(i.replace("\n","|"))
    except:
        target_students.append(" ")

op_df1["target_students"]=target_students
```

```python
prereq=[]
for i in urls:
    try:
        driver.get(i)
        pr=(driver.find_elements_by_class_name("egCard__item"))
        prereq.append(pr[1].find_element_by_tag_name("ul").text)
        
        #print("\n\n",i,"\n")
    except:
        prereq.append(" ")

prerequisites=[]
for i in prereq:
    try:
        prerequisites.append(i.replace("\n","|"))
    except:
        prerequisites.append(i)

op_df1["prerequisites"]=prerequisites
```
