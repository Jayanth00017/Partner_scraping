```python
urls=["https://teamtreehouse.com/library/practice-semantic-html","https://teamtreehouse.com/library/practice-structuring-and-grouping-content","https://teamtreehouse.com/library/practice-file-paths","https://teamtreehouse.com/library/practice-css-selectors","https://teamtreehouse.com/library/practice-css-box-model-basics","https://teamtreehouse.com/library/practice-media-queries","https://teamtreehouse.com/library/practice-basic-variables-input-output-in-javascript","https://teamtreehouse.com/library/practice-javascript-math-methods","https://teamtreehouse.com/library/practice-if-and-else-if-statements-in-javascript","https://teamtreehouse.com/library/practice-basic-javascript-functions","https://teamtreehouse.com/library/practice-javascript-loops-2","https://teamtreehouse.com/library/practice-object-literals-in-javascript"]
issue=["https://teamtreehouse.com/library/modern-css-layout-techniques","https://teamtreehouse.com/library/the-dom-and-es2015"]
```

```python
html_content=[]
soup=[]

for i in range (len(urls)):
    html_content.append(requests.get(urls[i]).text)
    soup.append(BeautifulSoup(html_content[i], "lxml"))
```

```python
len(soup)
```

```python
title=[]
for i in soup:
    title.append(i.title.text)
op_df["title"]=title
```

```python
description=[]
for i in soup:
    try:
        pt=(i.find(class_="grid-66 tablet-grid-60").find_all("p"))
        description.append(pt[1])
    except:
        print("Error")
op_df["description"]=description
```

```python
inst_1_name=[]
for i in soup:
    inst_1_name.append(i.find("h4").text)
op_df["instructor|1|name"]=inst_1_name
```

```python
level=[]
for i in soup:
    level.append(i.find("li",{"id" : "workshop-skill-level"}).text)
op_df["level"]=level
```

```python
inst_desig=[]
for i in soup:
    inst_desig.append("Teacher")
op_df["instructor|1|designation"]=inst_desig
```

```python
inst_1_bio=[]
for i in soup:
    inst_1_bio.append(i.find("div",class_="markdown-zone").text)
op_df["instructor|1|instructor_bio"]=inst_1_bio
```

```python
topics=[]
for i in soup:
        topics.append(i.find("ul",class_="tags actions").find("a").text)            
op_df["topics"]=topics
```

```python

learn_type=[]
for i in soup:
    if(i.text.find("Workshop"))>=1:
        learn_type.append("Workshop")
    else:
        learn_type.append("Course")
op_df["learn_type"]=learn_type
```

```python
inst_type=[]
for i in soup:
    inst_type.append("Self Paced")
op_df["instruction_type"]=inst_type
```

```python
inst_1_img=[]
for i in soup:
    px=(i.find(class_="instructor-avatar")['style'].split("("))
    inst_1_img.append(px[1])

op_df["instructor|1|instructor_image"]=inst_1_img
```

```python
duration=[]
duration_unit=[]
for i in soup:
    try:
        pt=(i.find("div",{"id" : "syllabus-title"}).find("h2").text.replace("\n","").split(" "))
        pt=(pt[4].split("-"))
        duration.append(pt[0])
        duration_unit.append(pt[1])
    except:
        pr=(i.find("div",{"id" : "workshop-title"}).find("h2").text.replace("\n","").split(" "))
        pr=(pr[4].split("-"))
        duration.append(pr[0])
        duration_unit.append(pr[1])
                
op_df["total_duration"]=duration
op_df["total_duration_unit"]=duration_unit       
```

```python
content=[]
for i in urls:
    tempro=[]
    driver.get(i)
    time.sleep(5)
    a=driver.find_elements_by_class_name("toggle-steps")
    for i in a:
        time.sleep(2)
        try:
            i.click()
        except:
            pass
    pri=(driver.find_elements_by_class_name("stage-meta"))
    for j in pri:
        prnt=(j.text.replace(":","").replace("questions","").replace("\n","|"))
        res = ''.join([i for i in prnt if not i.isdigit()])
        tempro.append(res)
    content.append(tempro)



op_df["content"]=content
```
