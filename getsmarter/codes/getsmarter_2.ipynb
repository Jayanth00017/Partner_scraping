```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import time
import dateutil.parser as parser

def getsmarter_politics_and_economics():

    course_links=[]
    certification_url='https://www.getsmarter.com/products/subject/politics-and-economics'
    base_url = 'https://www.getsmarter.com'
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'
    }

    r=requests.get(certification_url,headers=headers)
    soup=BeautifulSoup(r.content,"html.parser")
    course_list=soup.findAll("div",class_="product-block")
    for course in course_list:
      for link in course.findAll("a", href=True):
        course_links.append(base_url+link['href'])
    #print((course_links))

    title=[]
    short_desc=[]
    logo=[]
    univ_name=[]
    video_url=[]
    description=[]
    what_will_learn=[]
    skills_gained=[]
    price=[]
    skills2=[]
    target_students=[]
    learn_type=[]
    level=[]
    total_duration=[]
    total_duration_units=[]
    pricing_type=[]
    currency=[]
    language=[]
    accessibilities=[]
    delivery_method=[]
    instructor_name=[]
    instructor_name_2=[]
    instructor_desg=[]
    instructor_desg_2=[]
    instructor_img=[]
    instructor_img_2=[]
    instruction_type=[]
    batch_start_date=[]
    enrollment_end_date=[]
    recommended_effort_per_week=[]
    reviewer_name_1=[]
    reviewer_name_2=[]
    reviewer_name_3=[]
    reviewer_name_4=[]
    review_1=[]
    review_2=[]
    review_3=[]
    review_4=[]
    content=[]
    additional_payment_details=[]
    partner_course_url=[]



    for url in course_links:
      try:
        r=requests.get(url, headers=headers)
        soup=BeautifulSoup(r.content,"html.parser")
      except:
        time.sleep(3)

      #partner_course_url
      partner_course_url.append(url)


      #title
      titles=soup.find("h1",class_="heading-h1 course-name-heading").text.replace("\n","")
      title.append(titles)
    #print(title)

      #short_desc
      short=soup.find("p",class_="paragraph-default course-paragraph").text
      short_desc.append(short)
    #print(short_desc)

      #logo
      l=soup.find("img",class_="course-logo")['src']
      logo.append(l)

      #univ_name
      univ=soup.find("p",class_="paragraph-small-blue course-partner-name").text
      univ_name.append(univ)

      #description
      #desc=soup.find("div",class_="overview-text").findChild("div",class_="heading").text
      desc=soup.find("div",class_="heading").findChildren("p")
      description.append(desc)

      #language
      #lang=soup.find("div",class_="course-info-block")[1]
      langgg=soup.findAll("p",class_="heading-h5")[1].text
      language.append(langgg)

      #total_duration
      dur=soup.findAll("p",class_="heading-h5")[0].text.split()
      dura=dur[0]
      total_duration.append(dura)

      #total_duration_units
      total_duration_units.append(dur[1])

      #what_will_learn
      try:
        will=[]
        what = soup.findAll("div",class_="paragraph")
        for i in what:
          ii=i.find("span").text.replace(".","|")
          will.append(ii)

        what_will_learn.append(will)
      except:
        pass

      
      #instrcutor_name
      name=soup.find("p",class_="name").text
      instructor_name.append(name)

      #instrcutor_name_2
      try:
        name2=soup.findAll("p",class_="name")[1].text
        instructor_name_2.append(name2)
      except:
        pass

       #instrcutor_img
      try:
        if soup.find("img",class_="desktop-faculty-director-img") is None:
          img=soup.find("img",class_="desktop-convenor-img")['data-original']
          instructor_img.append(img)
        else:
          img2=soup.find("img",class_="desktop-faculty-director-img")['data-original']
          instructor_img.append(img2)
      except:
        pass

       #instrcutor_img_2
      try:
        if soup.find("img",class_="desktop-faculty-director-img") is None:
          img3=soup.findAll("img",class_="desktop-convenor-img")[1]['data-original']
          instructor_img_2.append(img3)
        else:
          img4=soup.findAll("img",class_="desktop-faculty-director-img")[1]['data-original']
          instructor_img_2.append(img4)
      except:
        pass



      #instrcutor_desg
      try:
        if soup.find("div",class_="faculty-director-text") is None:
          desg=soup.find("div",class_="convenor-text").findChildren("p")[2].text
          instructor_desg.append(desg)
        else:
          desg2=soup.find("div",class_="faculty-director-text").findChildren("p")[2].text
          instructor_desg.append(desg2)
      except:
        pass

      #instructor_desg_2
      try:
        if soup.find("div",class_="faculty-director-text") is None:
          desg3=soup.findAll("div",class_="convenor-text")[1].findChildren("p")[2].text
          instructor_desg_2.append(desg3)
        else:
          desg4=soup.findAll("div",class_="faculty-director-text")[1].findChildren("p")[2].text
          instructor_desg_2.append(desg4)
      except:
        pass

      #price
      try:
        cost=soup.findAll("p",class_="heading-h3")[1].text.split()
        costt=cost[0].replace("$","")
        price.append(costt)
      except:
        pass

      #Currency
      try:
        curr=cost[1]
        currency.append(curr)
      except:
        pass

      #content
      con=[]
      cont=soup.findAll("div",class_="module-content")
      content.append(cont)

      #target_students
      target=soup.findAll("div",class_="heading")[1].findChildren("p")[1].text.replace(".","|")
      target_students.append(target)
     
     #batch_start_date
      try:
       start=soup.find("p",class_="heading-h3").text
       date=parser.parse(start)

       batch_start_date.append(date.isoformat())
      except:
        pass

      #reviewer_name
      try:
        r_name=soup.find("div",class_="testimonial_name_title").findChild("p").text
        reviewer_name_1.append(r_name)
      except:
        pass

      #review
      try:
        rev=soup.find("div",class_="testimonial-text").findChild("p").text
        review_1.append(rev)
      except:
        pass

      #additional_payment_details
      additional_payment_details.append("You can pay your course fees before the course starts, or you may opt for a split payment plan on courses that are nine weeks and shorter. For courses 10 weeks and longer, there is a three-part payment plan available.")
      
      #enrollment_end_date
      try:
        enroll=soup.find("span",class_="paragraph-small-blue").text.replace("Enroll by","")
        end=parser.parse(enroll)
        enrollment_end_date.append(end.isoformat())
      except:
        pass

      #recommended_effort_per_week
      try:
        recommended=soup.findAll("p",class_="heading-h5")[2].text.replace("hours per week","")
        recommended_effort_per_week.append(recommended)
      except:
        pass

      #accessibilities
      accessibilities.append("Mobile,Desktop,Tablet")

      #delivery_method
      delivery_method.append("Online")

      #instruction_type
      instruction_type.append("Self Paced")

      #learn_type
      learn_type.append("Course")

      pricing_type.append("Paid")
 

      df=pd.DataFrame.from_dict({
                                            
                 
                                            'title': title,
                                            'learn_type' :learn_type,
                                            

                                            #'course_description': course_description,
                                            
                                            
                                            'description':description,
                                            'delivery_method': delivery_method,
                                            'instruction_type' :  instruction_type,
                                            'content': content,

                                            #'skills':skills_gained,
                                            #'content':content,
                                            'what_will_learn':what_will_learn,
                                            'target_students':target_students,
                                            

                                            'instructor|1|name':instructor_name,
                                            'instructor|1|designation':instructor_desg,
                                            'instructor|1|instructor_image':instructor_img,

                                            'instructor|2|name':instructor_name_2,
                                            'instructor|2|designation':instructor_desg_2,
                                            'instructor|2|instructor_image':instructor_img_2,
                                            

                                            'review|1|reviewer_name':reviewer_name_1,
                                            'review|1|review':review_1,

                                            'total_duration' : total_duration,
                                            'total_duration_unit':total_duration_units,
                                            'recommended_effort_per_week':recommended_effort_per_week,

                                            'batch|1|batch_start_date':batch_start_date,
                                            'enrollment_end_date':enrollment_end_date,
                                            'languages' : language,
                                            'short_description': short_desc,
                                            'accessibilities' : accessibilities,
                                            'pricing_type':pricing_type,  
                                            'currency':currency,
                                            'regular_price':price,
                                            'additional_pricing_details':additional_payment_details,
                                            'institute':univ_name,
                                            'partner_course_url': partner_course_url
                                            
     }, orient='index')
                                            
                            

      df = df.transpose()
      df.to_csv("getsmarter_politics_and_economics.csv", index=False)
                

                
if __name__ == '__main__':
    getsmarter_politics_and_economics()
```
